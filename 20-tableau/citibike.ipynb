```python
import pandas as pd
import datetime
```

```python
jan = pd.read_csv("JC-201901-citibike-tripdata.csv")
feb = pd.read_csv("JC-201902-citibike-tripdata.csv")
mar = pd.read_csv("JC-201903-citibike-tripdata.csv")
apr = pd.read_csv("JC-201904-citibike-tripdata.csv")
may = pd.read_csv("JC-201905-citibike-tripdata.csv")
jun = pd.read_csv("JC-201906-citibike-tripdata.csv")
```

```python
jan.head()
```

```python
feb.head()
```

```python
#merge tables
dfs=[jan, feb, mar,apr, may, jun]
merge=pd.concat(dfs)
merge.info()
merge.count()
```

```python
merge.head()
```

```python
# prepare for data merge between two csv
new = merge["starttime"].str.split(" ", n = 1, expand = True)  
# making separate first name column from new data frame 
merge["Date"]= new[0] 
merge.head()
merge.tail()
```

```python
#output dataset
merge.info()
merge.to_csv("Trip History 2019Jan-June.csv", index=False, header=True)
```
